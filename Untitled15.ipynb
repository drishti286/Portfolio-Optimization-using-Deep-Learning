```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import StandardScaler
from scipy.optimize import minimize
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, LSTM, Conv1D, MaxPooling1D, Flatten, Input, MultiHeadAttention, LayerNormalization, Add

# Constants
TICKERS = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA"]
START_DATE = "2014-01-01"
END_DATE = "2024-01-01"
TRAIN_YEARS = 8
TEST_YEARS = 2

def calculate_rsi(data, window=14):
    delta = data['Close'].diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

def generate_signals(data):
    signals = pd.DataFrame(index=data.index)
    signals['Daily return'] = data['Close'].pct_change()
    signals['Volatility'] = signals['Daily return'].rolling(window=5).std()
    signals['Volume'] = data['Volume']
    signals['rsi'] = calculate_rsi(data)
    return signals[['Daily return', 'Volatility', 'Volume', 'rsi']].dropna()

data = {}
for ticker in TICKERS:
    stock_data = yf.download(ticker, start=START_DATE, end=END_DATE)
    data[ticker] = generate_signals(stock_data)

def prepare_data(stock_data, train_years, test_years):
    train_size = int(len(stock_data) * train_years / (train_years + test_years))
    train_data = stock_data[:train_size]
    test_data = stock_data[train_size:]

    scaler = StandardScaler()
    scaled_train = scaler.fit_transform(train_data)
    scaled_test = scaler.transform(test_data)

    X_train, y_train = [], []
    for i in range(60, len(scaled_train)):
        X_train.append(scaled_train[i-60:i, :-1])
        y_train.append(scaled_train[i, -1])

    X_test, y_test = [], []
    for i in range(60, len(scaled_test)):
        X_test.append(scaled_test[i-60:i, :-1])
        y_test.append(scaled_test[i, -1])

    return (
        np.array(X_train), np.array(y_train),
        np.array(X_test), np.array(y_test),
        scaler
    )

def build_mlp_model(input_shape):
    model = Sequential([
        Input(shape=input_shape),
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def build_cnn_model(input_shape):
    model = Sequential([
        Input(shape=input_shape),
        Conv1D(filters=64, kernel_size=3, activation='relu'),
        MaxPooling1D(pool_size=2),
        Conv1D(filters=32, kernel_size=3, activation='relu'),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(50, activation='relu'),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input
from tensorflow.keras import regularizers

def build_lstm_model(input_shape):
    inputs = Input(shape=input_shape)
    x = LSTM(units=100, return_sequences=True, kernel_regularizer=regularizers.l2(0.01))(inputs)
    x = Dropout(0.2)(x)
    x = LSTM(units=50, return_sequences=False, kernel_regularizer=regularizers.l2(0.01))(x)
    x = Dropout(0.2)(x)
    x = Dense(64, activation='relu')(x)
    outputs = Dense(1)(x)
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model


def build_transformer_model(input_shape, num_heads=4, ff_dim=128, dropout_rate=0.2):
    input_layer = Input(shape=input_shape)
    x = MultiHeadAttention(num_heads=num_heads, key_dim=input_shape[-1])(input_layer, input_layer)
    x = Dropout(dropout_rate)(x)
    x = Add()([input_layer, x])
    x = LayerNormalization(epsilon=1e-6)(x)

    ff = Dense(ff_dim, activation='relu')(x)
    ff = Dense(input_shape[-1])(ff)
    x = Dropout(dropout_rate)(ff)
    x = Add()([x, ff])
    x = LayerNormalization(epsilon=1e-6)(x)

    x = Flatten()(x)
    output = Dense(1)(x)

    model = Model(inputs=input_layer, outputs=output)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def portfolio_optimization(expected_returns, cov_matrix):
    n_assets = len(expected_returns)

    def objective(weights):
        portfolio_return = np.dot(weights, expected_returns)
        portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
        sharpe_ratio = portfolio_return / portfolio_volatility
        return -sharpe_ratio

    constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})
    bounds = tuple((0.05, 0.6) for _ in range(n_assets))
    init_guess = [1.0 / n_assets] * n_assets

    result = minimize(objective, init_guess, method='SLSQP', bounds=bounds, constraints=constraints)
    return result.x

portfolio_returns_dict = {"MLP": [], "CNN": [], "LSTM": [], "Transformer": []}

for ticker in TICKERS:
    stock_data = data[ticker]
    X_train, y_train, X_test, y_test, scaler = prepare_data(stock_data, TRAIN_YEARS, TEST_YEARS)

    # Reshape input data
    X_train_cnn = X_train.reshape((X_train.shape[0], X_train.shape[1], X_train.shape[2]))
    X_test_cnn = X_test.reshape((X_test.shape[0], X_test.shape[1], X_test.shape[2]))

    X_train_flat = X_train.reshape(X_train.shape[0], -1)
    X_test_flat = X_test.reshape(X_test.shape[0], -1)

    # Train MLP Model
    mlp_model = build_mlp_model((X_train_flat.shape[1],))
    mlp_model.fit(X_train_flat, y_train, epochs=20, batch_size=32, verbose=0)
    predicted_mlp = mlp_model.predict(X_test_flat)

    # Train CNN Model
    cnn_model = build_cnn_model((X_train_cnn.shape[1], X_train_cnn.shape[2]))
    cnn_model.fit(X_train_cnn, y_train, epochs=20, batch_size=32, verbose=0)
    predicted_cnn = cnn_model.predict(X_test_cnn)

    # Train LSTM Model
    lstm_model = build_lstm_model((X_train.shape[1], X_train.shape[2]))
    lstm_model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=0)
    predicted_lstm = lstm_model.predict(X_test_cnn)

    # Train Transformer Model
    transformer_model = build_transformer_model((X_train.shape[1], X_train.shape[2]))
    transformer_model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=0)
    predicted_transformer = transformer_model.predict(X_test_cnn)

    # Process predictions for all models
    for model_name, predicted in zip(["MLP", "CNN", "LSTM", "Transformer"], [predicted_mlp, predicted_cnn, predicted_lstm, predicted_transformer]):
        predicted_prices = scaler.inverse_transform(
            np.concatenate((np.zeros((len(predicted), X_test.shape[2])), predicted), axis=1)
        )[:, -1]
        actual_prices = scaler.inverse_transform(
            np.concatenate((np.zeros((len(y_test), X_test.shape[2])), y_test.reshape(-1, 1)), axis=1)
        )[:, -1]

        actual_returns = np.diff(actual_prices) / actual_prices[:-1]
        predicted_returns = np.diff(predicted_prices) / predicted_prices[:-1]
        positions = np.where(predicted_returns > 0, 1, -1)
        portfolio_returns_dict[model_name].append(positions * actual_returns)

for model_name, portfolio_returns in portfolio_returns_dict.items():
    portfolio_returns = np.array(portfolio_returns)
    cov_matrix = np.cov(portfolio_returns)
    expected_returns = np.mean(portfolio_returns, axis=1)
    optimized_weights = portfolio_optimization(expected_returns, cov_matrix)
    weighted_returns = np.dot(portfolio_returns.T, optimized_weights)

    cumulative_portfolio_returns = np.cumsum(weighted_returns)
    plt.figure(figsize=(12, 6))
    plt.plot(cumulative_portfolio_returns, label=f'{model_name} Portfolio Cumulative Returns')
    plt.title(f'{model_name} Aggregated Long-Short Portfolio Returns')
    plt.xlabel('Time')
    plt.ylabel('Cumulative Returns')
    plt.legend()
    plt.show()

    cumulative_return = cumulative_portfolio_returns[-1]
    sharpe_ratio = np.mean(weighted_returns) / np.std(weighted_returns) * np.sqrt(252)
    max_drawdown = np.min(cumulative_portfolio_returns - np.maximum.accumulate(cumulative_portfolio_returns))

    print(f"{model_name} Model:")
    print(f"Cumulative Return: {cumulative_return:.2f}")
    print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
    print(f"Max Drawdown: {max_drawdown:.2f}\n")
```
